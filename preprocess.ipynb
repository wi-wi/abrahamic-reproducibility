# Preprocessing Pipeline
Implements Appendix C: cleaning, normalization, uncertainty propagation.


```python
import pandas as pd
import numpy as np
from pathlib import Path
df = pd.read_csv('proxies_babylonian.csv')
df
```

## Sort & align time

```python
df = df.sort_values('year').reset_index(drop=True)
df['dt'] = df['year'].diff().fillna(0).astype(int)
df
```

## Normalize to [0,1]

```python
def scale(s):
    mn, mx = s.min(), s.max()
    return (s - mn) / (mx - mn)
for col in ['x_obs','v_proxy','s_proxy','gamma_proxy']:
    df[col + '_norm'] = scale(df[col])
df.to_csv('proxies_babylonian_normalized.csv', index=False)
df
```
